```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('SampleSuperstore.csv')
data.head(5)
```

```python
data.shape
```

```python
data.describe()
```

```python
sns.jointplot(x='Sales',y='Quantity', data=data)
```

```python
data.isna().sum()
```

```python
data.isnull().any()
```

```python
corr=data.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
```

```python
plt.scatter(x=data['Sales'],y=data['Profit'],c='black')
plt.title("Sale & Profit")
plt.xlabel('Total Sale')
plt.ylabel("Profit")
plt.show()
```

```python
data.hist(figsize=(14,14),bins=50)
plt.show()
```

## Total Profit & Sales

```python
plt.figure(figsize=(12,7))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data['State'],order=(data['State'].value_counts().head(30)).index)
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(8,12))
data['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Sub-Category',hue='Region',data=data)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data['Segment'])
plt.show()
```

```python
sns.countplot(data['Ship Mode'])
plt.show()
```
